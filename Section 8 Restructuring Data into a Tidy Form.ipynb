```python
import pandas as pd
movie=pd.read_csv('data/movie.csv')
actor=movie[['movie_title','actor_1_name','actor_2_name','actor_3_name','actor_1_facebook_likes','actor_2_facebook_likes'\
            ,'actor_3_facebook_likes']]
actor.head()

#change column name
def change_col_name(col_name):
    col_name=col_name.replace('_name','')
    if 'facebook' in col_name:
        fb_idx=col_name.find('facebook') #找facebook的位置
        col_name=col_name[:5]+col_name[fb_idx-1:]+col_name[5:fb_idx-1]
        
    return col_name

actor2=actor.rename(columns=change_col_name)
actor2.head()

stubs=['actor','actor_facebook_likes']
actor2_tidy=pd.wide_to_long(actor2,stubnames=stubs,i='movie_title',j='actor_num',sep='_')
actor2_tidy    


df=pd.read_csv('data/stackme.csv')
df2=df.rename(columns={'a1':'group1_a1','b2':'group1_b2','d':'group2_a1','e':'group2_b2'})
pd.wide_to_long(df2,stubnames=['group1','group2'],i=['State','Country','Test'],j='Label',suffix='.+',sep='_')
#suffix default is /d, 去看regular expression
```

```python
#invert stacked data
import pandas as pd
usecol_func=lambda x:'UGDS_' in x or x=='INSTNM'
college=pd.read_csv('data/college.csv',index_col='INSTNM',usecols=usecol_func)
college.head()

college_stacked=college.stack(dropna=True)
college_stacked.head(10)

#变回去
college_stacked.unstack()

#metl and pivot
college2=pd.read_csv('data/college.csv',usecols=usecol_func)
college_melt=college2.melt(id_vars='INSTNM',var_name='Race',value_name='Percentage')
college_melt.head()

#use pivot to invert this previous result:
meltd_inv=college_melt.pivot(index='INSTNM',columns='Race',values='Percentage')
meltd_inv.head() #index 的 order变了

college2_replication=meltd_inv.loc[college2['INSTNM'],college2.columns[1:]].reset_index()
college2_replication.head()

college.stack() #这个有2个index 一个是INSTNM 一个是USDS..

#如果在unstack里面什么参数都不给，他默认用最里面的index，也就是USDS作为cuolumn，这种情况下
#最外面的index是0,然后-1,-2向里面的index内推，所以你要用outermost column using level=0
college.stack().unstack(0)

#T or Transpose()
college.T
```

```python
#Replicate pivot_table with a groupby aggregation
import pandas as pd
import numpy as np
flights=pd.read_csv('data/flights.csv')
fp=flights.pivot_table(index='AIRLINE',columns='ORG_AIR',values='CANCELLED',aggfunc='sum',fill_value=0).round(2)
#fill_value 吧NAN 弄成0, aggfunc default is mean
fp

#use groupby to do the same
fg=flights.groupby(['AIRLINE','ORG_AIR'])['CANCELLED'].agg(np.sum)
fg_unstack=fg.unstack('ORG_AIR',fill_value=0)
fg_unstack
#fp.equals(fg_unstack)

#more complex pivot table
flights.pivot_table(index=['AIRLINE','MONTH'],columns=['ORG_AIR','CANCELLED'],values=['DEP_DELAY','DIST'],\
                    aggfunc=[np.sum,np.mean],fill_value=0)

flights.groupby(['AIRLINE','MONTH','ORG_AIR','CANCELLED'])['DEP_DELAY','DIST'].agg([np.sum,np.mean]).\
unstack(['ORG_AIR','CANCELLED']).fillna(0).swaplevel(0,1,axis='columns')

```

```python
#Renaming axis levels fro eash reshaping
import pandas as pd
college=pd.read_csv('data/college.csv')
cg=college.groupby(['STABBR','RELAFFIL'])\
           ['UGDS','SATMTMID'].agg(['size','min','max']).head(10)
#注意columns有2个level但是都没有名字
cg=cg.rename_axis(['AGG_COLS','AGG_FUNCS'],axis='columns')
#现在每个columns都有名字，就可以stack了
cg.stack('AGG_FUNCS').head()
#但是按照默认的，stack以后的level会在最里面，我们可以用swaplevel
cg.stack('AGG_FUNCS').swaplevel('AGG_FUNCS','STABBR',axis='index').head()

#we can continue to make sure of the axis level name by sorting levels with sort_index method
cg.stack('AGG_FUNCS').swaplevel('AGG_FUNCS','STABBR',axis='index')\
.sort_index(level='RELAFFIL',axis='index')
```

```python
#Tidying when mutiple variables are stored as columns names
import pandas as pd
weightlifting=pd.read_csv('data/weightlifting_men.csv')
weightlifting
wl_melt=weightlifting.melt(id_vars='Weight Category',var_name='sex_age',value_name='Qual Total')
wl_melt.head(10)
#str.split Series.str.split(pat=None, n=-1, expand=False)[source]
#pat : str, optional String or regular expression to split on. If not specified, split on whitespace.
#expand : bool, default False 
#If True, return DataFrame/MultiIndex expanding dimensionality.
#If False, return Series/Index, containing lists of strings.
sex_age=wl_melt['sex_age'].str.split(expand=True)
sex_age.columns=['Sex','Age Group']
sex_age['Sex']=sex_age['Sex'].str[0] #就是截图35的第一个，再赋值回去，相当于left(str,1)

#use pd.concat 
wl_cat_total=wl_melt[['Weight Category','Qual Total']]
wl_tiny=pd.concat([sex_age,wl_cat_total],axis=1) #当axis = 1(columns)的时候，concat就是行对齐，然后将不同列名称的两张表合并
#https://blog.csdn.net/stevenkwong/article/details/52528616
wl_tiny
```

```python
#Tidy when mutltiple variables are stored as column values
inspections=pd.read_csv('data/restaurant_inspections.csv',parse_dates=['Date'])
inspections.head(10)
#two variable name and date
inspections.set_index(['Name','Date','Info']).unstack('Info')
#make the index levels into columns with reset_index
#DataFrame.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
#col_level : int or str, default 0
#If the columns have multiple levels, determines which level the labels are inserted into. 
#By default it is inserted into the first level.
insp_tity=inspections.set_index(['Name','Date','Info']).unstack('Info').reset_index(col_level=-1)
insp_tity.columns=insp_tity.columns.droplevel(0).rename(None)
insp_tity
```

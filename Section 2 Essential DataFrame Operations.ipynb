```python
college=pd.read_csv('data\college.csv',index_col='INSTNM')
college_ugds=college.filter(like='UGDS_')
college_ugds.head()
college_ugds_round=(college_ugds+1).round(2) 

#Compare missing values
college_ugds.isnull().sum()

#Transposing the direction of a DataFrame operations
college_ugds.count() #equal to   college_ugds.count(axis=0) or college_ugds.count(axis='index')
college_ugds.median(axis='index')
college_ugds.count(axis='columns')
college_ugds.cumsum(axis=1)
```

```python
#Determining college campus diversity
college_ugds.isnull()\
            .sum(axis=1)\
            .sort_values(ascending=False)
college_ugsd_new=college_ugds.dropna(how='all')
college_ugsd_new
college_ugsd_new.ge(.15).sum(axis=1).sort_values(ascending=False) #test all values >=0.15, and sum by row for those boolean value
college_ugds.loc[['Regency Beauty Institute-Austin','Central Texas Beauty College-Temple']]
(college_ugsd_new>0.001).all(axis=1).any()          #retrun to true if all are true, then return to true if any is ture
```
